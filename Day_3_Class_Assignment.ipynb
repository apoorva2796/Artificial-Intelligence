```python
import tensorflow as tf

import os
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.metrics import confusion_matrix
import warnings
from sklearn.metrics import accuracy_score
warnings.filterwarnings('ignore')

path = r"C:\Users\Administrator\Desktop\Folders\Term-3\AI\Dataset"
os.chdir(path)
data = pd.read_csv("sonar.csv")
data.head()
```

```python
dict_1 = {'Rock': 0,'Mine': 1} 
data.Class = [dict_1[item] for item in data.Class] 
```

def rename(val):
    if val == 'Rock':
        return 0
    else:
        return 1
    
data['Label'] = data['Class'].apply(rename)

```python
data.head()
```

```python
#defining the perceptron

#define placeholder for features and labels
X_input = data[data.columns.difference(['Class'])]

Y_label = data[['Class']]
```

```python
Y_label.head()
```

```python
X_input.shape
```

```python
data.shape
```

```python
#define placeholder for features and labels
x = tf.placeholder(tf.float32,[None,60]) # inputs/features -x1, x2
#Labels
y = tf.placeholder(tf.float32,[None,1]) #Labels


# weights and bias
weight = tf.Variable(tf.zeros([60,1]))

bias = tf.Variable(tf.zeros([1]))

#input function

I = tf.add(tf.matmul(x,weight), bias)

#activation function

output = tf.sigmoid(I)
```

```python
#define the cost or loss function amd the optimizier

individual_loss = tf.nn.sigmoid_cross_entropy_with_logits(labels=y, logits=output)
loss = tf.reduce_mean(individual_loss)

gdo = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = gdo.minimize(loss)

#initialize the variables

init = tf.global_variables_initializer()
```

```python
sess = tf.Session()
sess.run(init)
```

```python
sess.run(bias)
```

```python
for i in range(550000):
    final_train = sess.run(train, feed_dict={x:X_input, y:Y_label})
```

```python
sess.run(bias)
```

```python
result = sess.run(output,feed_dict={x:X_input})

result
```

```python
y_pred = result.round()

y_pred
```

```python
accuracy_score(Y_label,y_pred)
```

```python
confusion_matrix(Y_label, y_pred)
```
