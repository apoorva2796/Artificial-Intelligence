```python
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.applications.vgg16 import preprocess_input
from keras.applications.vgg16 import decode_predictions
from keras.applications.vgg16 import VGG16

from keras.applications.resnet50 import ResNet50
from keras.preprocessing import image
from keras.applications.resnet50 import preprocess_input
```

```python
model = VGG16()
```

```python
#load an image from file

image = load_img(r'C:\Users\Administrator\Desktop\2-keras-cnn-catdog\dataset\single_prediction\dog1.jpg', target_size=(224,224))
image
```

```python
#convert the image pixel to numpy array

image = img_to_array(image)

#reshape data for model
image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))

#prepare the image for VGG
image = preprocess_input(image)
```

```python
# predict the probability across all output classes
result = model.predict(image)
# convert the probabilities to class labels
label = decode_predictions(result)
# retrieve the most likely result, e.g. highest probability
label = label[0][0]
# print the classification
print('%s (%.2f%%)' % (label[1], label[2]*100))
```
