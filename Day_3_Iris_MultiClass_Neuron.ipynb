```python
import tensorflow as tf

import os
import numpy as np
import pandas as pd 
from pandas import get_dummies
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.metrics import confusion_matrix
import warnings
from sklearn.metrics import accuracy_score
warnings.filterwarnings('ignore')

from sklearn.datasets import load_iris

path = r"C:\Users\Administrator\Desktop\Folders\Term-3\AI\Dataset"
os.chdir(path)
data = pd.read_csv("Iris_dataset.csv")
data.head()
```

```python
X_input = data.drop(labels='Species', axis=1)

Y_label = data.iloc[:,4]
```

```python
y_dummies = pd.get_dummies(Y_label)

y_hot = y_dummies.values
```

```python
#defining the perceptron

#define placeholder for features and labels
x = tf.placeholder(tf.float32,[None,4])

y = tf.placeholder(tf.float32,[None,3])


#weights and bias

weight = tf.Variable(tf.truncated_normal([4,3],stddev=0.1))

bias = tf.Variable(tf.truncated_normal([1],stddev=0.1))

#input function

netinput = tf.add(tf.matmul(x,weight), bias)

#activation function

output = tf.nn.softmax(netinput)
```

```python
#define the cost or loss function amd the optimizier

loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(labels=y, logits=output))


#Optimizer
gdo = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = gdo.minimize(loss)

#initialize the variables

init = tf.global_variables_initializer()
```

```python
sess = tf.Session()
sess.run(init)
```

```python
for i in range(5000):
    final_train = sess.run(train, feed_dict={x:X_input, y:y_hot})
```

```python
predict = sess.run(output, feed_dict={x:X_input})

predict
```

```python
result = sess.run(tf.arg_max(predict,1))

result
```

```python
labhot = sess.run(tf.arg_max(y_hot,1))

labhot
```

```python
accuracy_score(result,labhot)
```

```python
confusion_matrix(result,labhot)
```
